```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/cleaned_data_tashkent.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

X_train=train_set.drop('price', axis=1)
y_train=train_set['price']

X_num = X_train.drop(['location', 'district'], axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder, StandardScaler

num_pipeline=Pipeline([
                      
                       ('std_scaler', StandardScaler() )
])
```

```python
from sklearn.compose import ColumnTransformer

num_att=list(X_num)
cat_att=['location', 'district']

full_pipeline = ColumnTransformer([
                                   ('num', num_pipeline, num_att),
                                   ('cat', OrdinalEncoder(), cat_att)
])
```

```python
X_prepared = full_pipeline.fit_transform(X_train) 
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
RF_model.fit(X_prepared, y_train)
```

```python
X_test=test_set.drop('price', axis=1)
y_test=test_set['price']
```

```python
X_test_prepared=full_pipeline.fit_transform(X_test)
```

```python
y_predicted=RF_model.predict(X_test_prepared)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
rf_mse = mean_squared_error(y_test, y_predicted)
rf_rmse = np.sqrt(rf_mse)
r2=r2_score(y_test, y_predicted)
print(f'RMSE: {rf_rmse}\nR2: {r2}')
```
